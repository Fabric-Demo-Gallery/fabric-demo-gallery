# Bronze Layer — Ingest Raw Education Data
Reads CSV files from lakehouse and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
```

```python
# Ingest faculty
df_faculty = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/faculty.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df_faculty.write.mode('overwrite').format('delta').saveAsTable('bronze_faculty')
print(f'Bronze faculty: {spark.table("bronze_faculty").count()} rows')
```

```python
# Ingest students
df_students = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/students.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df_students.write.mode('overwrite').format('delta').saveAsTable('bronze_students')
print(f'Bronze students: {spark.table("bronze_students").count()} rows')
```

```python
# Ingest enrolments
df_enrolments = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/enrolments.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df_enrolments.write.mode('overwrite').format('delta').saveAsTable('bronze_enrolments')
print(f'Bronze enrolments: {spark.table("bronze_enrolments").count()} rows')
```

```python
# Ingest assessments
df_assessments = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/assessments.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df_assessments.write.mode('overwrite').format('delta').saveAsTable('bronze_assessments')
print(f'Bronze assessments: {spark.table("bronze_assessments").count()} rows')
```

```python
print('Bronze ingestion complete.')
print(f'  Faculty     : {df_faculty.count():>7,}')
print(f'  Students    : {df_students.count():>7,}')
print(f'  Enrolments  : {df_enrolments.count():>7,}')
print(f'  Assessments : {df_assessments.count():>7,}')
```
