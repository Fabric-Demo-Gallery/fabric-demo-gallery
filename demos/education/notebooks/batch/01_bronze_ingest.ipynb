# Bronze Layer — Ingest Raw Education Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

ingestion_batch_id = str(uuid.uuid4())
```

```python
# Ingest students
df = (
    spark.read.format('csv').option('header', True).option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/students.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df.write.mode('overwrite').format('delta').saveAsTable('bronze_students')
print(f'bronze_students: {df.count()} rows')
```
