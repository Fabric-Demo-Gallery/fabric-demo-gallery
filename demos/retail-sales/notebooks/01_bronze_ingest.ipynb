# Bronze Layer — Ingest Raw Retail Data
Reads CSV files from lakehouse `Files/landing/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
```

```python
# Ingest POS transactions
df_transactions = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('Files/landing/pos_transactions.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_transactions.write.mode('overwrite').format('delta').saveAsTable('bronze_pos_transactions')
print(f'Bronze POS transactions: {df_transactions.count()} rows')
```

```python
# Ingest product catalog
df_products = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('Files/landing/products.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_products.write.mode('overwrite').format('delta').saveAsTable('bronze_products')
print(f'Bronze products: {df_products.count()} rows')
```
