# Bronze Layer — Ingest Raw Hospitality Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
```

```python
# Ingest properties
df_properties = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/properties.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_properties.write.mode('overwrite').format('delta').saveAsTable('bronze_properties')
print(f'Bronze properties: {spark.table("bronze_properties").count()} rows')
```

```python
# Ingest guests
df_guests = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/guests.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_guests.write.mode('overwrite').format('delta').saveAsTable('bronze_guests')
print(f'Bronze guests: {spark.table("bronze_guests").count()} rows')
```

```python
# Ingest bookings
df_bookings = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/bookings.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_bookings.write.mode('overwrite').format('delta').saveAsTable('bronze_bookings')
print(f'Bronze bookings: {spark.table("bronze_bookings").count()} rows')
```

```python
# Ingest reviews
df_reviews = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/reviews.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_reviews.write.mode('overwrite').format('delta').saveAsTable('bronze_reviews')
print(f'Bronze reviews: {spark.table("bronze_reviews").count()} rows')
```

```python
print('Bronze ingestion complete.')
print(f'  Properties : {df_properties.count():>7,}')
print(f'  Guests     : {df_guests.count():>7,}')
print(f'  Bookings   : {df_bookings.count():>7,}')
print(f'  Reviews    : {df_reviews.count():>7,}')
```
