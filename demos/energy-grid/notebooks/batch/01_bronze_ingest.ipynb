# Bronze Ingest — Energy Grid (Lakehouse)

Ingests CSV data into Bronze Delta tables in the Lakehouse.
This is the Lakehouse variant (for ML scenarios, not KQL).

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
print(f'Bronze ingest batch: {batch_id}')
```

```python
# Grid sensors
df = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/grid_sensors.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df.write.mode('overwrite').format('delta').saveAsTable('bronze_grid_sensors')
print(f'Bronze grid sensors: {df.count()} rows')
```

```python
# Power events
df = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/power_events.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df.write.mode('overwrite').format('delta').saveAsTable('bronze_power_events')
print(f'Bronze power events: {df.count()} rows')
```

```python
# Renewable generation
df = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/renewable_generation.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)
df.write.mode('overwrite').format('delta').saveAsTable('bronze_renewable_generation')
print(f'Bronze renewable generation: {df.count()} rows')
```

```python
print('\nBronze ingest complete')
```
