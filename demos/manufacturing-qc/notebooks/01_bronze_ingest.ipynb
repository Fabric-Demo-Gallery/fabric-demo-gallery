# Bronze Layer — Ingest Raw Sensor & Production Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

ingestion_batch_id = str(uuid.uuid4())
```

```python
# Ingest sensor readings
df_sensors = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/sensor_readings.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)

df_sensors.write.mode('overwrite').format('delta').saveAsTable('bronze_sensor_readings')
print(f'Bronze sensor readings: {df_sensors.count()} rows')
```

```python
# Ingest production batches
df_batches = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/production_batches.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)

df_batches.write.mode('overwrite').format('delta').saveAsTable('bronze_production_batches')
print(f'Bronze production batches: {df_batches.count()} rows')
```

```python
# Ingest equipment catalog
df_equipment = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/equipment_catalog.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)

df_equipment.write.mode('overwrite').format('delta').saveAsTable('bronze_equipment_catalog')
print(f'Bronze equipment catalog: {df_equipment.count()} rows')
```
