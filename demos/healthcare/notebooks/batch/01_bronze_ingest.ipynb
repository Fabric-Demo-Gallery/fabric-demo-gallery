# Bronze Layer — Ingest Raw Healthcare Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

ingestion_batch_id = str(uuid.uuid4())
```

```python
# Ingest staff catalog
df_staff = (
    spark.read.format('csv').option('header', True).option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/staff_catalog.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_staff.write.mode('overwrite').format('delta').saveAsTable('bronze_staff_catalog')
print(f'Bronze staff catalog: {spark.table("bronze_staff_catalog").count()} rows')
```

```python
# Ingest patient admissions
df_admissions = (
    spark.read.format('csv').option('header', True).option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/patient_admissions.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_admissions.write.mode('overwrite').format('delta').saveAsTable('bronze_patient_admissions')
print(f'Bronze patient admissions: {spark.table("bronze_patient_admissions").count()} rows')
```

```python
# Ingest clinical records (vitals)
df_clinical = (
    spark.read.format('csv').option('header', True).option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/clinical_records.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_clinical.write.mode('overwrite').format('delta').saveAsTable('bronze_clinical_records')
print(f'Bronze clinical records: {spark.table("bronze_clinical_records").count()} rows')
```
