# Bronze Layer — Ingest Raw Financial Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

ingestion_batch_id = str(uuid.uuid4())
```

```python
# Ingest customers
df_customers = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/customers.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_customers.write.mode('overwrite').format('delta').saveAsTable('bronze_customers')
print(f'Bronze customers: {df_customers.count()} rows')
```

```python
# Ingest accounts
df_accounts = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/accounts.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_accounts.write.mode('overwrite').format('delta').saveAsTable('bronze_accounts')
print(f'Bronze accounts: {df_accounts.count()} rows')
```

```python
# Ingest transactions
df_transactions = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/transactions.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('ingestion_batch_id', lit(ingestion_batch_id))
)
df_transactions.write.mode('overwrite').format('delta').saveAsTable('bronze_transactions')
print(f'Bronze transactions: {df_transactions.count()} rows')
```
