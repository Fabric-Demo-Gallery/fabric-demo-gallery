# Bronze Layer — Ingest Raw Media & Telecom Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
```

```python
# Ingest content catalog
df_content = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/content_catalog.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_content.write.mode('overwrite').format('delta').saveAsTable('bronze_content_catalog')
print(f'Bronze content catalog: {spark.table("bronze_content_catalog").count()} rows')
```

```python
# Ingest subscribers
df_subs = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/subscribers.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_subs.write.mode('overwrite').format('delta').saveAsTable('bronze_subscribers')
print(f'Bronze subscribers: {spark.table("bronze_subscribers").count()} rows')
```

```python
# Ingest viewing history
df_views = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/viewing_history.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_views.write.mode('overwrite').format('delta').saveAsTable('bronze_viewing_history')
print(f'Bronze viewing history: {spark.table("bronze_viewing_history").count()} rows')
```

```python
# Ingest ad impressions
df_ads = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/ad_impressions.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_ads.write.mode('overwrite').format('delta').saveAsTable('bronze_ad_impressions')
print(f'Bronze ad impressions: {spark.table("bronze_ad_impressions").count()} rows')
```

```python
print('Bronze ingestion complete.')
print(f'  Content Catalog : {df_content.count():>8,}')
print(f'  Subscribers     : {df_subs.count():>8,}')
print(f'  Viewing History : {df_views.count():>8,}')
print(f'  Ad Impressions  : {df_ads.count():>8,}')
```
