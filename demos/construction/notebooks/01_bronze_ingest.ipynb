# Bronze Layer — Ingest Raw Construction Data
Reads CSV files from lakehouse `{{DATA_SOURCE_PATH}}/` and writes Delta tables with metadata columns.

```python
from pyspark.sql.functions import current_timestamp, input_file_name, lit
import uuid

batch_id = str(uuid.uuid4())
```

```python
# Ingest subcontractors
df_subs = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/subcontractors.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_subs.write.mode('overwrite').format('delta').saveAsTable('bronze_subcontractors')
print(f'Bronze subcontractors: {df_subs.count()} rows')
```

```python
# Ingest projects
df_projects = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/projects.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_projects.write.mode('overwrite').format('delta').saveAsTable('bronze_projects')
print(f'Bronze projects: {df_projects.count()} rows')
```

```python
# Ingest tasks
df_tasks = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/tasks.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_tasks.write.mode('overwrite').format('delta').saveAsTable('bronze_tasks')
print(f'Bronze tasks: {df_tasks.count()} rows')
```

```python
# Ingest cost ledger
df_costs = (
    spark.read.format('csv')
    .option('header', True)
    .option('inferSchema', True)
    .load('{{DATA_SOURCE_PATH}}/cost_ledger.csv')
    .withColumn('ingestion_timestamp', current_timestamp())
    .withColumn('source_file', input_file_name())
    .withColumn('batch_id', lit(batch_id))
)

df_costs.write.mode('overwrite').format('delta').saveAsTable('bronze_cost_ledger')
print(f'Bronze cost ledger: {df_costs.count()} rows')
```

```python
print('Bronze ingestion complete.')
print(f'  Subcontractors : {df_subs.count():>7,}')
print(f'  Projects       : {df_projects.count():>7,}')
print(f'  Tasks          : {df_tasks.count():>7,}')
print(f'  Cost Ledger    : {df_costs.count():>7,}')
```
